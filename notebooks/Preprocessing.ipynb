<h1>Preprocessing</h1>

```python
import pandas as pd
import datetime
```

```python
data_food = pd.read_excel('data/food.xlsx')
```

```python
data_food = data_food.sort_values(['sigUf', 'ano', 'mês'])
```

```python
def trimAllColumns(df):
    """
    Trim whitespace from ends of each value across all series in dataframe
    """
    trimStrings = lambda x: x.strip() if type(x) is str else x
    return df.applymap(trimStrings)
```

```python
data_food = trimAllColumns(data_food)
```

```python
grupos = set(data_food['desGrupoVenda'].tolist())
```

```python
collection_uf = set(data_food['sigUf'].tolist())
```

```python
df = pd.DataFrame()
```

```python
df['UF'] = None
df['DATA'] = None
df['VENDA'] = None
for grupo in grupos:
    df[grupo] = None
```

```python
data_food = data_food[data_food['sigUf'] == 'RJ']
```

```python
collection_ano = list(range(2010,2018))
collection_mes = list(range(1,13))
```

```python
for ano in collection_ano:     
    row_dict = {}
    venda = 0
    for mes in collection_mes:
        for index, row_food in data_food.iterrows(): 
            if ano == row_food['ano'] and mes == row_food['mês']:
                venda = venda  + round(row_food['Total'],2)
                row_dict[row_food['desGrupoVenda']] = round(row_food['Total'],2)
                row_dict['DATA'] = datetime.datetime(row_food['ano'], row_food['mês'], 1)
                row_dict['UF'] = row_food['sigUf']
        if row_dict:
            row_dict['VENDA'] = round(venda,2)
            df = df.append(row_dict, ignore_index=True) 
            venda = 0
```

```python
df.fillna(0)
```

```python
writer = pd.ExcelWriter('data/food-rj.xlsx')
df.to_excel(writer,'sp')
writer.save()
```
