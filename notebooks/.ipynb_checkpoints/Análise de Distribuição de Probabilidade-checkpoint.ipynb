# Análise de Distribuição de Probabilidade

```r
library(grid)
library(ggplot2)
```

```r
dados.vendas <- read.csv("data/food-sp.csv", header=TRUE)
dados.vendas = as.data.frame(dados.vendas)
dados.vendas[is.na(dados.vendas)] <- 0
dados.vendas$PROJETO.CAFÉ[dados.vendas$PROJETO.CAFÉ == 0] <- mean(dados.vendas$PROJETO.CAFÉ)
dados.vendas$BRINDE[dados.vendas$BRINDE == 0] <- mean(dados.vendas$BRINDE)
dados.vendas = dados.vendas[, c(
                                'GELADO.E.MILK.SHAKE', 
                                'SANDUICHE', 
                                'BEBIDA', 
                                'ACOMPANHAMENTO',
                                'ADICIONAL', 
                                'PROJETO.CAFÉ',
                                'BRINDE',
                                'PRATO'
                               )] 

```

```r
library(fitdistrplus)
```

```r
descdist(dados.vendas$GELADO.E.MILK.SHAKE, boot = 500)
```

```r
variavel = dados.vendas$GELADO.E.MILK.SHAKE

fit.lnorm <- fitdist(variavel, "lnorm")
fit.weibull <- fitdist(variavel, "weibull")
fit.norm <- fitdist(variavel, "norm")

#datacol='purple',
denscomp(list(fit.lnorm, fit.weibull, fit.norm), 
         main="Análise de Distribuição de Probabilidade - GELADO.E.MILK.SHAKE",
         xlab="Variável = GELADO.E.MILK.SHAKE",
         ylab="Densidade",
         plotstyle="ggplot", 
         legendtext=c("LogNormal", "Weibull", "Normal"))
```

```r
rbind(c("LOGNORMAL", "WEIBULL", "NORMAL"))
aic = rbind(c(fit.lnorm$aic, fit.weibull$aic, fit.norm$aic))
aic
min(aic)
```
